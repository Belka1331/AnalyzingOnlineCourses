Объедините датасеты (кроме students.scv) в один общий для дальнейшей работы. Отдельный датасет создайте для расчета среднего возраста студентов курсов (отдельный датасет со students.scv нужен, чтобы в основном не потерять значения в основном датасете из-за того, что не по всем студентам есть анкетные данные о дате рождения).


```python
#** Можно добавить скриншоты таблиц courses и course_students, показать сколько строк в них, и проверить сколько строк после соединения. 
#Тут можно добавить более подробное объяснение того, чем мы тут занимаемся. Ex: объединим дата сеты courses и courses_contents под столбцам id & course_id соответственно
#dataset=pd.merge(courses,course_contents, left_on='id', right_on='course_id') - invalid
dataset=pd.merge(course_contents,courses, left_on='course_id', right_on='id')
del(dataset['Unnamed: 0'])
del(dataset['id'])
#dataset.head(10)
#соединим progress
prog_merged=pd.merge(progresses,progress_phases,left_on='id', right_on='progress_id')
prog_merged=prog_merged.drop(columns='id')
#prog_merged.head(10)
dataset=pd.merge(dataset,prog_merged, on=['course_id','module_number','lesson_number'])
dataset['start_date'] = pd.to_datetime(
    dataset['start_date'], format='ISO8601'
)
dataset['start_date']=dataset['start_date'].dt.floor('s').dt.tz_localize(None)
dataset['finish_date'] = pd.to_datetime(
    dataset['finish_date'], format='ISO8601'
)
dataset['finish_date']=dataset['finish_date'].dt.floor('s').dt.tz_localize(None)
dataset=dataset.astype({'is_homework':bool})
dataset.head()
```

```python
dur_df_main = dataset.groupby(['student_id', 'title', 'module_number']).agg(
                                                             start_date_len = pd.NamedAgg(column='start_date', aggfunc=lambda x: len(x)), 
                                                             start_date_list = pd.NamedAgg(column='start_date', aggfunc=lambda x: list(x)), # lessons_done
                                                             start_date_min = pd.NamedAgg(column='start_date', aggfunc=lambda x: x.min()),# start_min
                                                             start_date_max =pd.NamedAgg (column='start_date', aggfunc=lambda x: x.max()), # finish_max
                                                             dur_start =pd.NamedAgg(column='start_date', aggfunc=lambda x: (max(x) - min(x)).days), # days of learning 
                                                             dur_finish = pd.NamedAgg(column='finish_date', aggfunc=lambda x: (max(x) - min(x)).days)).\
                                                             sort_values(by='dur_start', ascending=False).\
                                                             query('(dur_start > 60) & (dur_finish > 60)')
```

```python
mydur = dataset.groupby(['student_id', 'title', 'module_number']).agg(
                                                             started_lessons = pd.NamedAgg(column='start_date', aggfunc=lambda x: len(x)), 
                                                             finished_lessons = ('finish_date', lambda x: x.notna().sum()), 
                                                             start_date_list = ('start_date', lambda x: list(x)),
                                                             start_min = ('start_date', lambda x: x.min()),
                                                             finish_max = ('finish_date', lambda x: x.max() if x.notna().any() else pd.NaT))
mydur['dif']=mydur['finish_max']-mydur['start_min']
mydur['dif']=mydur['dif'].dt.days
mydur=mydur.sort_values(by='dif',ascending=False).query('dif>=30')
```

```python
mydur
```

```python
sns.histplot(data=mydur,x='dif')
plt.title('Гистограмма распределения разницы в днях между минимальной\n\
        и максимальной датой прохождения всеми студентами уроков в модулях\n\
        по курсам за весь период')
plt.xlabel('дней')
plt.ylabel('Кол-во модулей')
```
