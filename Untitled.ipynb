```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
%matplotlib inline
```

Codebook
courses.csv содержит следующие значения:

     id – идентификатор курса
     title – название курса
     field – сфера, к которой относится курс


students.csv содержит следующие значения:

     id – идентификатор студента
     city – город студента
     birthday – день рождения студента


course_contents.csv содержит следующие значения:

     course_id – идентификатор курса
     module_number – номер модуля
     module_title – название модуля
     lesson_number – номер урока
     lesson_title – название урока
     lesson_token – токен урока
     is_video – наличие видео (true/false)
     is_homework – наличие домашней работы (true/false)


progresses.csv содержит следующие значения:

     id – идентификатор прогресса
     student_id – идентификатор студента
     course_id – идентификатор курса


progress_phases.csv содержит следующие значения:

     progress_id – идентификатор прогресса
     module_number – номер модуля
     lesson_number – номер урока
     status – статус прохождения урока
     start_date – дата начала
     finish_date – дата окончания

```python
# загузка фреймов данных
try:
    courses = pd.read_csv('courses.csv')
    #в students.csv имеем неккоретное значение в строке 4605 - 0235-05-31, может есть и другие
    students = pd.read_csv('students.csv')
    course_contents = pd.read_csv('course_contents.csv')
    progresses = pd.read_csv('progresses.csv')
    progress_phases = pd.read_csv('progress_phases.csv')
except:
    print('Ошибка. Файл не найден')
else:
    print('Файл прочитан')
```

Объедините датасеты (кроме students.scv) в один общий для дальнейшей работы. Отдельный датасет создайте для расчета среднего возраста студентов курсов (отдельный датасет со students.scv нужен, чтобы в основном не потерять значения в основном датасете из-за того, что не по всем студентам есть анкетные данные о дате рождения).


```python
#** Можно добавить скриншоты таблиц courses и course_students, показать сколько строк в них, и проверить сколько строк после соединения. 
#Тут можно добавить более подробное объяснение того, чем мы тут занимаемся. Ex: объединим дата сеты courses и courses_contents под столбцам id & course_id соответственно
#dataset=pd.merge(courses,course_contents, left_on='id', right_on='course_id') - invalid
dataset=pd.merge(course_contents,courses, left_on='course_id', right_on='id')
del(dataset['Unnamed: 0'])
del(dataset['id'])
#dataset.head(10)
#соединим progress
prog_merged=pd.merge(progresses,progress_phases,left_on='id', right_on='progress_id')
prog_merged=prog_merged.drop(columns='id')
#prog_merged.head(10)
dataset=pd.merge(dataset,prog_merged, on=['course_id','module_number','lesson_number'])
dataset['start_date'] = pd.to_datetime(
    dataset['start_date'], format='ISO8601'
)
dataset['start_date']=dataset['start_date'].dt.floor('s').dt.tz_localize(None)
dataset['finish_date'] = pd.to_datetime(
    dataset['finish_date'], format='ISO8601'
)
dataset['finish_date']=dataset['finish_date'].dt.floor('s').dt.tz_localize(None)
dataset.head(10)
```
